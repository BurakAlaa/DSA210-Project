```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the data file
file_path = 'processed_data/CS300_data.txt'  
data = pd.read_csv(file_path)

# Step 2: Extract the columns of interest
# Make sure the column names match the ones in your file
date_column = 'DATE'
total_minutes_column = 'TOTAL MINUTES STUDIED'

# Step 3: Process the data
# Ensure the date column is parsed correctly (optional: convert to datetime)
data[date_column] = pd.to_datetime(data[date_column], format='%d.%m.%Y', errors='coerce')

# Remove any rows with invalid dates or missing data
data = data.dropna(subset=[date_column, total_minutes_column])

# Step 4: Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data[date_column], data[total_minutes_column], marker='o', linestyle='-', label='Total Study Minutes')
plt.title('Total Study Minutes per Day (CS300)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Study Minutes', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(visible=True, which='major', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Define the file names and directory
file_names = [
    'CS300_data.txt',
    'CS303_data.txt',
    'CS308_data.txt',
    'DSA210_data.txt',
    'MATH204_data.txt'
]
directory = 'processed_data'

# Loop through each file
for file_name in file_names:
    file_path = os.path.join(directory, file_name)
    
    # Read the file
    data = pd.read_csv(file_path)
    
    # Extract the columns of interest
    date_column = 'DATE'
    total_minutes_column = 'TOTAL MINUTES STUDIED'
    time_blocks = ['09.00 - 12.00', '12.00 - 17.00', '17.00 - 21.00', '21.00 - 00.00', '00.00 - 03.00']
    location_column = 'STUDY LOCATION'
    
    # Ensure the date column is parsed correctly
    data[date_column] = pd.to_datetime(data[date_column], format='%d.%m.%Y', errors='coerce')
    data = data.dropna(subset=[date_column, total_minutes_column])  # Drop invalid rows
    
    # ---------------- Total Study Minutes by Location (IC vs Room) ----------------
    # Filter and group by study locations
    location_totals = data.groupby(location_column)[total_minutes_column].sum()
    location_totals = location_totals.loc[['IC', 'Room']]  # Keep only IC and Room
    
    # Plot location data
    plt.figure(figsize=(8, 5))
    location_totals.plot(kind='bar', color=['skyblue', 'orange'], edgecolor='black')
    plt.title(f'Total Study Minutes by Location (IC vs Room) - {file_name.split("_")[0]}', fontsize=16)
    plt.xlabel('Study Location', fontsize=12)
    plt.ylabel('Total Minutes Studied', fontsize=12)
    plt.xticks(rotation=0)
    plt.grid(visible=True, axis='y', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()
```
